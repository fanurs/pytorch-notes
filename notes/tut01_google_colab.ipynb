# CPU vs GPU

PyTorch tensors are essentially like NumPy arrays, but they are designed in a way that can be run on a GPU. This is especially useful for training deep learning models, which involves intensive matrix computation.

On Google Colab, it's easy to choose between CPU or GPU. Simply go to "Runtime", select "Change runtime type".

image.png

To select CPU, choose "None"; to select GPU, choose "GPU".

image.png

After that, simply "Connect to a hosted runtime".

image.png

We are now ready to run Python code.

```python
import torch
torch.cuda.device_count()
```

If you have connected to a GPU runtime, you should see `torch.cuda.device_count()` equals to `1`. If you have connected to a CPU runtime, then it should be `0`.

Whether you are using CPU or GPU, the PyTorch APIs are the same. In the upcoming tutorials, we will see how easy it is to switch to GPU when necessary. Otherwise, we will be using CPU runtime.
